```python
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
car_dekhoo.drop(columns=['seller_type'],inplace=True)
car_dekhoo
```

```python
car_dekhoo['year']=2024-car_dekhoo['year']
car_dekhoo
```

```python
car_dekhoo['company']=car_dekhoo['name'].apply(lambda x:x.split()[0])
```

```python
car_dekhoo
```

```python
car_dekhoo.drop(columns=['name'],inplace=True)
car_dekhoo
```

```python
car=LinearRegression()
```

```python
car_dekhoo=pd.get_dummies(car_dekhoo,drop_first=True)
car_dekhoo
```

```python
x=car_dekhoo.drop(columns=['selling_price'])
y=car_dekhoo['selling_price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=2,test_size=0.2)
car.fit(x_train,y_train)
```

```python
y_pred=car.predict(x_test)
```

```python
pd.DataFrame({'actual':y_test,'predicted':y_pred})
```

```python
from sklearn.metrics import r2_score,mean_squared_error
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=20)
```

```python
y_pred=l.predict(x_test)
y_pred
```

```python
# polynomial process start here
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly=PolynomialFeatures(degree=4)
x_poly=poly.fit_transform(x)
x_train,x_test,y_train,y_test=train_test_split(x_poly,y,test_size=0.2,random_state=20)
```

```python
l=LinearRegression()
l.fit(x_train,y_train)
```

```python
y_pred=l.predict(x_test)
y_pred
```

```python
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
real.info()
```

```python
real.isna().sum()
```

```python
real.drop(columns=['No'],inplace=True)
real
```

```python
sns.heatmap(real.corr(),annot=True,cmap='YlGnBu')
```

```python
x=real.drop(columns=['Y house price of unit area'])
y=real['Y house price of unit area']
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2)
x_poly=poly.fit_transform(x)
x_poly.shape
```

```python
# fromula how we get 28 columns 
# (n+d)! / (d! * (n+d-d)!)
# n=no of x varibaled=degrees

```

```python
x_train,x_test,y_train,y_test=train_test_split(x_poly,y,test_size=0.25,random_state=10)
plr=LinearRegression()

```

```python
plr.fit(x_train,y_train)
y_pred=plr.predict(x_test)
```

```python
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```
